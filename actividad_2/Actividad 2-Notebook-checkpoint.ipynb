```python
import numpy as np
import pandas as pd
import csv
```

```python
datos_read = []
with open('./estaciones_bici.csv') as File:
    reader = csv.reader(File, delimiter=';', quotechar=',', quoting=csv.QUOTE_MINIMAL)
    for row in reader:
        datos_read.append(row)
datos_read
```

```python
# Construir un DataFrame con diccionarios. Como si fueran registros con diccionarios Python
df_datos = pd.DataFrame(datos_read)
# fijarse que se crea el df y los valores sin datos se quedan en NaN
df_datos                    #Data Frame creado para limpieza y consultas
```

```python
df_datos['libres'] = df_datos['free'] + df_datos['available']
```
